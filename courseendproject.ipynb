```python
#handle Imbalanced Data
not_fully_paid0=df[df['not.fully.paid']==0]
not_fully_paid1=df[df['not.fully.paid']==1]
```

```python
#resample
from sklearn.utils import resample
df_sampled_data=resample(not_fully_paid1,replace=True,n_samples=8045)
```

```python
df_sampled_data.shape
```

```python
new_df=pd.concat([not_fully_paid0,df_sampled_data])
```

```python
from sklearn.utils import shuffle
new_df=shuffle(new_df)
```

```python
new_df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in new_df.columns:
    if new_df[i].dtypes=='object':
        new_df[i]=le.fit_transform(new_df[i])
```

```python
x=new_df[['credit.policy', 'purpose', 'int.rate', 'installment','fico', 'revol.bal', 'revol.util','inq.last.6mths','pub.rec']]
```

```python
x.shape
```

```python
y=new_df[['not.fully.paid']]
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#apply standardScaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.fit_transform(xtest)
```

```python
# build the predictive models
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping

model =Sequential()
```

```python
#first Layer
model.add(Dense(19,activation='relu',input_shape=(9,)))
#dropout is used to handle overfitting
model.add(Dropout(0.3))
#Add second Layer
model.add(Dense(10,activation='relu'))
model.add(Dropout(0.3))
#output layer
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#early stopping

early_stop=EarlyStopping(monitor='val_loss',min_delta=0.01,mode='min',verbose=1,patience=10)
```

```python
history=model.fit(xtrain,ytrain,epochs=50,batch_size=256,validation_data=(xtest,ytest),callbacks=[early_stop])
```

```python
model.evaluate(xtest,ytest)
```

```python
model.evaluate(xtrain,ytrain)
```

```python
#plot the accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['train','test'],loc='best')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.legend(['train','test'],loc='best')
plt.show()
```
